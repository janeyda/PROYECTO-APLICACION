### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# Informe final


El objetivo de este análisis realizado es poder evaluar el desempeño de las cuatro tiendas del Sr. Juan, para poder decidir qué tienda de su cadena Alura Store debe vender para iniciar un nuevo emprendimiento, considerando ingresos totales, volumen de ventas, productos más y menos vendidos, calificaciones promedio de los clientes y costos de envío.
Teniendo en cuenta que el costo de envío es asumido por el cliente, no es un factor muy relevante, ya que no afecta las ganancias de don Juan, pero sí puede influir en la decisión de compra.

1. Ingresos totales y volumen de ventas

En los análisis y visualizaciones realizadas se pudo conocer que la tienda con mayores ingresos es la Tienda 1, con un total de 1,150,880,400, seguida muy de cerca por la Tienda 2, con un total de 1,116,343,500. La Tienda 3 muestra ingresos similares; sin embargo, la tienda con menos ingresos es la Tienda 4, con 1,038,375,700.
Si solo se tienen en cuenta los ingresos, la Tienda 4 sería la que se debería vender; sin embargo, aún no se puede determinar esta decisión sin considerar las otras características que se analizarán a continuación.

2. Productos más y menos vendidos

La Tienda 1 tiene una fuerte demanda en productos como el Armario.
La Tienda 2 tiene una fuerte demanda en el producto Iniciando en programación.
La Tienda 4 tiene productos más vendidos, pero también presenta productos con muy baja rotación.
Por último, la Tienda 3 muestra ventas más equilibradas que las otras tiendas, pero sin un producto con una diferencia marcada frente a las demás tiendas.

3. Calificación promedio de los clientes

En cuanto a satisfacción del cliente, la Tienda 3 obtiene la mejor calificación promedio (4.05), seguida de la Tienda 2 (4.04), la Tienda 4 (4.0) y, finalmente, la Tienda 1 (3.98).
Aunque la Tienda 4 no tiene una mala calificación, tampoco se destaca frente a las demás.

4. Costo promedio de envío

En las gráficas se puede observar que la Tienda 4 tiene el menor costo de envío; sin embargo, teniendo en cuenta que este costo es pagado por el cliente, no afecta la rentabilidad del Sr. Juan y no es suficiente para compensar sus menores ingresos.

**Recomendación**

Se recomienda que el Sr. Juan venda la Tienda 4, ya que es la que muestra el menor nivel de ingresos y no destaca significativamente en la satisfacción del cliente, a diferencia de la Tienda 2 y la Tienda 3, que presentan ingresos sólidos y una buena calificación en la satisfacción del cliente.
Aunque la Tienda 1 tiene el nivel de satisfacción del cliente más bajo, es la que genera mayores ingresos, por lo cual no es recomendable venderla. Vender la Tienda 4 permitiría al Sr. Juan concentrar recursos y esfuerzos en las tiendas con mayor rentabilidad y mejor posicionamiento en el mercado.
