```python
fig, axs = plt.subplots(2,2,figsize=(12,8))

fig.subplots_adjust(hspace=0.9, wspace=0.5)

fig.suptitle('DistribuciÃ³n de Churn segÃºn variables categÃ³ricas', fontsize=14)

def churn_plot(col, ax):
    tabla = pd.crosstab(empresa[col], empresa['Churn'], normalize='index')*100
    tabla.plot(kind='bar', ax=ax)

    ax.set_title(f'Churn por {col}')
    ax.set_ylabel('% clientes')
    ax.set_xlabel(col)


    ax.set_ylim(0,100)


    ax.grid()
churn_plot('gender', axs[0,0])
churn_plot('Contract', axs[0,1])
churn_plot('PaymentMethod', axs[1,0])
churn_plot('InternetService', axs[1,1])

plt.show()
```

ðŸ”¹ **IntroducciÃ³n**

El objetivo de este anÃ¡lisis es identificar los factores que influyen en la evasiÃ³n de clientes (Churn) en Telecom X. La evasiÃ³n de clientes representa un problema importante para la empresa, ya que implica pÃ©rdida de ingresos y mayores costos de adquisiciÃ³n de nuevos clientes.

Los datos fueron importados desde un archivo JSON y convertidos en un DataFrame de Pandas.

Pasos realizados:

-NormalizaciÃ³n de datos

Se normalizaron las columnas anidadas (customer, phone, internet, account) para obtener un dataset plano y usable.

-RevisiÃ³n de tipos de datos

- Se detectaron valores nulos en Churn.

  Charges.Monthly Ã— tenure

-CreaciÃ³n de nueva variable

Cuentas_Diarias = Charges.Monthly / 30

Esto permitiÃ³ analizar el gasto diario de los clientes.


ðŸ”¹ **AnÃ¡lisis Exploratorio de Datos**


Los grÃ¡ficos mostraron que una parte relevante de los clientes ha cancelado el servicio, lo que confirma que el churn es un problema real para la empresa.


-Churn segÃºn variables categÃ³ricas
